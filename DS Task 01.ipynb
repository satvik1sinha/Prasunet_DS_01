```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Generate a sample dataset
sample_size = 1000
gender_distribution = np.random.choice(['Male', 'Female', 'Other'], sample_size, p=[0.49, 0.49, 0.02])
age_distribution = np.random.randint(18, 80, sample_size)  # Ages between 18 and 80

# Create a DataFrame
data = pd.DataFrame({
    'Gender': gender_distribution,
    'Age': age_distribution
})
```

```python
# Plot the gender distribution
plt.figure(figsize=(10, 6))
data['Gender'].value_counts().plot(kind='bar', color=['blue', 'pink', 'green'])
plt.title('Gender Distribution in Sample Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Plot the age distribution
plt.figure(figsize=(10, 6))
data['Age'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution in Sample Population')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
